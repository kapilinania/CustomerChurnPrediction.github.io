##features Selection

```python
#features Selection
x=data.drop(['Churn'], axis=1)
y=data['Churn']
```

```python
x
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.experimental import enable_hist_gradient_boosting
```

##Data Spliting


##Data Scaling



```python
from sklearn.preprocessing import StandardScaler
scaler_data = StandardScaler()
fitx_data = scaler_data.fit_transform(x)
```

```python
print(fitx_data,y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(fitx_data, y, test_size=0.25, random_state=42)
x_train, x_test, y_train, y_test
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x_train
```

```python
from sklearn.ensemble import HistGradientBoostingClassifier


# Instantiate the HistGradientBoostingClassifier
modelGradient = HistGradientBoostingClassifier()

# Train the model
modelGradient.fit(x_train, y_train)

# Make predictions on the test set
y_pred = modelGradient.predict(x_test)

# Evaluate the model's performance
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

# Print predicted values
print("Predicted Values:\n", y_pred)

print("Accuracy:", accuracy)
print("Classification Report:\n", report)
print("Confusion Matrix:\n", conf_matrix)
```

```python
plt.figure(figsize=(6, 6))

ax2 = sns.distplot(data['Churn'], hist=False, color="r", label="Actual Value")
sns.distplot(y_pred, hist=False, color="b", label="Fitted Values" , ax=ax2)

plt.title('Actual vs Fitted Values for Churn  Blue is prediction')
plt.xlabel('Churn')
plt.ylabel('Proportion of Churn')

plt.show()
plt.close()
```

```python
from sklearn.ensemble import RandomForestClassifier

# Instantiate the RandomForestClassifier
model = RandomForestClassifier()

# Train the model
model.fit(x_train, y_train)

# Make predictions on the test set
y_pred = model.predict(x_test)

# Evaluate the model's performance
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

# Print predicted values
print("Predicted Values:\n", y_pred)

print("Accuracy:", accuracy)
print("Classification Report:\n", report)
print("Confusion Matrix:\n", conf_matrix)
```
